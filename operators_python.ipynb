```python
#comparision operators->,<,>=,<=,==!=(return True or False)
a=5
b=10
c=10
d=20
print(a>b)
print(b<c)
print(b>=c)
print(b==c)
print(b!=d)
```

```python
#logical operator - and, or, not
# and - this operator checking all conditions are true result in true rest cases failed
#or - thos operator check either one condition true then result true rest cases failed
a=10
b=20
c=20
d=60
print(a>b and b==c)
print(a<b and c<d)
print(a>b or b==c)
print(a>b or b!=c)
print(not a>b or b!=c)
```

```python
#membership operator - checking members are present in that sequence or collection - in, not in
names=["dheeraj","vikas","divya","alok"]
print("alok" in names)
print("chandra" in names)
print("dheeraj" not in names)
print("vikas" in names)
```

```python
#identity operators - identification of values - is, is not
a=10000
b=10000
print(id(a))
print(id(b))
print(a is b)
```
